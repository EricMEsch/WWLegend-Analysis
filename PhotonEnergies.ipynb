```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = '/home/eric/sim/WWLegend/Archive/Neutrontaggerruns/Warwick without ArgonModerator/'
energies =  np.loadtxt(file_path + "Foil On PMMA/PhotonEnergies.csv", dtype=float, delimiter=',', skiprows=5, unpack=True)
energies =  1239.84193 / (1e6 *energies) # convert energy to wavelength in nm
```

```python
fig, ax = plt.subplots()

n_bins = 1000
rangeBins = (0, 100000)
ax.hist(energies)
ax2 = ax.twinx()
ax2.hist(energies, cumulative=True, density=True, color="#DDAA33", histtype='step')

ax2.set_ylabel('Cumulative', color="#DDAA33")
ax.set_xlabel('Wavelength in nm')
ax.set_ylabel('Hist')
ax.set_title('Energie from Neutrons entering PMMA')
ax.legend(loc= "upper right")
#ax.set_xscale("log")
#ax.set_yscale("log")
ax.vlines(ymin=0,ymax=4e6, x=(300,500), colors="red")
ax2.grid()

#plt.savefig("/home/eric/sim/WWLegend/Plots/Energies.png")
plt.show()
```
